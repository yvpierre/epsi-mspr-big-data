```python
import pandas as pd

df = pd.read_csv("sources/valeurs_trimestrielles.csv", sep=';')

# Modify to get columns containing "- T"
selected_columns = [col for col in df.columns if col.startswith('Libell√©') or '-T' in col]

df_clean = df[selected_columns]

# Filtering out rows starting with 'Codes' in 'Libell√©' column
df_clean = df_clean[~df_clean['Libell√©'].str.startswith('Codes')]

# Modifying the first column values
df_clean.iloc[:, 0] = df_clean.iloc[:, 0].apply(lambda x: x.replace('Taux de ch√¥mage localis√© par d√©partement - ', ''))

# Further filtering out rows starting with 'Taux de ch√¥mage' in 'Libell√©' column
df_clean = df_clean[~df_clean['Libell√©'].str.startswith('Taux de ch√¥mage')]

# Saving the DataFrame to CSV without index
df_clean.to_csv("output/res.csv", index=False)

df_clean
```

```python
grouped_columns = df_clean.groupby(lambda x: x.split('-')[0], axis=1)

df_avg = grouped_columns.apply(lambda x: x.mean(axis=1))

df_avg
```

```python
df = df_clean.melt(id_vars=["Libell√©"], var_name="Quarter", value_name="Value")

df['Year'] = df['Quarter'].str.split('-').str[0]

df['Value'] = pd.to_numeric(df['Value'], errors='coerce')

df_grouped = df.groupby(['Libell√©', 'Year'])['Value'].mean().reset_index()

df_pivoted = df_grouped.pivot(index='Libell√©', columns='Year', values='Value').reset_index()

print(df_pivoted)
df_pivoted.round(2).to_csv("res2.csv")
```

```python

# Convert values to percentages
df_years_avg = df_pivoted.div(100)
# Calculate row-wise sum
df_years_avg['Average'] = df_years_avg.sum(axis=1)

# Transpose the dataframe
df_transposed = df_years_avg.transpose()

# Rename the index to 'Year'
df_transposed.index.name = 'Year'

# Reset index to move 'Year' to a column
df_transposed.reset_index(inplace=True)

# Calculate the mean of all values from columns other than the first one
average_all_except_first = df_transposed.iloc[:, 1:].mean(axis=1)

# Create a new DataFrame with 'Year' and 'Average' columns
average_df = pd.DataFrame({'Year': df_transposed['Year'], 'Average': average_all_except_first})

print(average_df)



average_df.to_csv("output/years_avg_chomage.csv")
```
